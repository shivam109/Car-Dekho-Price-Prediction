As we can see all the datasets have same column and data in each column is in json format. Lets Parse each json column to extract relevant information using json library and create new relevant column for e.g fuel_type, milege, seats, etc. we will do this for each column in the combined dataset and the preprocess those columns. we will parse only columns that will be helpful in price prediction.



```python
import ast

# Function to evaluate literal expression
def eval_literal(expression):
    try:
        # Evaluate the literal expression
        return ast.literal_eval(expression)
    except (SyntaxError, ValueError):
        return {}

# Apply the function to 'new_car_detail' column
df_combined['new_car_detail'] = df_combined['new_car_detail'].apply(eval_literal)

# Extract information from 'new_car_detail'
df_combined['Fuel_Type'] = df_combined['new_car_detail'].apply(lambda x: x.get('ft', ''))
df_combined['Transmission'] = df_combined['new_car_detail'].apply(lambda x: x.get('transmission', ''))
df_combined['Owner'] = df_combined['new_car_detail'].apply(lambda x: x.get('owner', ''))
df_combined['OEM'] = df_combined['new_car_detail'].apply(lambda x: x.get('oem', ''))
df_combined['Model'] = df_combined['new_car_detail'].apply(lambda x: x.get('model', ''))
df_combined['Model_Year'] = df_combined['new_car_detail'].apply(lambda x: x.get('modelYear', ''))
df_combined['Price'] = df_combined['new_car_detail'].apply(lambda x: x.get('price', ''))
```

```python
# Drop the original 'new_car_detail' column
df_combined.drop('new_car_detail', axis=1, inplace=True)
```

```python
import pandas as pd
import json

# Assuming df_combined is your combined dataset

# Extract information from 'new_car_overview' column
def extract_overview_info(row):
    try:
        data = eval(row['new_car_overview'])
        top_data = data['top']
        info_dict = {item['key'].replace(' ', '_'): item['value'] for item in top_data}
        return info_dict
    except (KeyError, TypeError):
        return {
            'Registration_Year': '',
            'Insurance_Validity': '',
            'Fuel_Type': '',
            'Seats': '',
            'Kms_Driven': '',
            'RTO': '',
            'Ownership': '',
            'Engine_Displacement': '',
            'Transmission': '',
            'Year_of_Manufacture': ''
        }

# Apply the function to create new columns
df_combined[['Registration_Year', 'Insurance_Validity', 'Fuel_Type', 'Seats', 'Kms_Driven', 'RTO',
              'Ownership', 'Engine_Displacement', 'Transmission', 'Year_of_Manufacture']] = \
    df_combined.apply(extract_overview_info, axis=1, result_type='expand')


```

```python
# Drop the original 'new_car_overview' column
df_combined = df_combined.drop(columns=['new_car_overview'])
```

```python
import pandas as pd
import ast

# Extracting data from 'new_car_feature' column
df_combined['new_car_feature'] = df_combined['new_car_feature'].apply(lambda x: {} if pd.isna(x) else ast.literal_eval(x))

# Creating new columns
df_combined['Feature_Heading'] = df_combined['new_car_feature'].apply(lambda x: x.get('heading', ''))
df_combined['Top_Features'] = df_combined['new_car_feature'].apply(lambda x: ', '.join([item['value'] for item in x.get('top', [])]))

# Extracting data from 'data' within 'new_car_feature'
df_combined['Category_Heading'] = df_combined['new_car_feature'].apply(lambda x: x.get('data', [])[0].get('heading', '') if x.get('data') else '')
df_combined['Category_SubHeading'] = df_combined['new_car_feature'].apply(lambda x: x.get('data', [])[0].get('subHeading', '') if x.get('data') else '')
df_combined['Category_Features'] = df_combined['new_car_feature'].apply(lambda x: ', '.join([item['value'] for item in x.get('data', [])[0].get('list', [])]) if x.get('data') else '')
```

```python
# Dropping the original 'new_car_feature' column if needed
df_combined = df_combined.drop(columns=['new_car_feature'])
```

```python
import ast

# Function to convert string representation to dictionary using ast.literal_eval
def convert_to_dict(specs_str):
    try:
        return ast.literal_eval(specs_str)
    except (SyntaxError, ValueError):
        return {}

# Convert 'new_car_specs' column to actual dictionaries
df_combined['new_car_specs'] = df_combined['new_car_specs'].apply(convert_to_dict)

# Function to extract values from 'top' or 'data' list based on 'key'
def extract_value(specs_dict, key):
    for section in ['top', 'data']:
        items = specs_dict.get(section, [])
        if isinstance(items, list):
            for item in items:
                if isinstance(item, dict) and item.get('key') == key:
                    return item.get('value')
        elif isinstance(items, dict) and items.get('key') == key:
            return items.get('value')
    return None

# Extract and create new columns
df_combined['Mileage'] = df_combined['new_car_specs'].apply(lambda x: extract_value(x, 'Mileage'))
df_combined['Engine'] = df_combined['new_car_specs'].apply(lambda x: extract_value(x, 'Engine'))
df_combined['Max_Power'] = df_combined['new_car_specs'].apply(lambda x: extract_value(x, 'Max Power'))
df_combined['Torque'] = df_combined['new_car_specs'].apply(lambda x: extract_value(x, 'Torque'))
df_combined['Seats'] = df_combined['new_car_specs'].apply(lambda x: extract_value(x, 'Seats'))
```

```python
# drop original car_specs column
df_combined = df_combined.drop(columns='new_car_specs')
```
